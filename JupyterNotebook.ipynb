### Feature Engineering

#### 1st - On Date of Journey feature


```python

df['Date'] = df['Date_of_Journey'].str.split('/').str[0] # to get only date
df['Month'] = df['Date_of_Journey'].str.split('/').str[1] # to get only month
df['Year'] = df['Date_of_Journey'].str.split('/').str[2] # to get only year
```

```python
df.head()

# (Successfully added as can be seen last 3 columns are added in our dataset)
```

```python
#Above splitting of date,month,year can also be done using lambda function(Always a good practise to google & find out a better way)

'''
df['Date'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
df['Month'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
df['Year'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
'''
```

```python
 df.info()
 
# since it can be seen that date, month, year are in object so we need to convert it into int
```

```python
# Converting date, month, year into int

df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()

# (Successfully changed the datatype)
```

```python
# Since we have Derived feature from Date of Journey col so we dont need it anymore. So dropping it

df.drop('Date_of_Journey',  axis = 1, inplace = True)
df.head(2)

# (Succeddfully removed)
```

```python
# Always remember while doing feature engineering we should always focus on similar features (Which we want to do again & again)
# As of now we are dealing with time data so the next col. I choose for feature engineering will be Arrival time 
```

#### 2nd - On Arrival time feature

```python
# Now we dont want to keep date part in arrival time. We only need to keep time. So we will drop date part in it.

df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_Time'].head()

# (Successfully removed date part which was present in this feature.)
```

```python
# Now splitting arrival time into arrival hour and arrival minute

df['Arrival_Hour'] = df['Arrival_Time'].apply(lambda x:x.split(':')[0]) #done splitting using lambda
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1] #done splitting normally

df.head()
```

```python
df.info()

# (Will convert Arrival hour and Arrival minute to int datatype)
```

```python
# Converting Arrival hour and minute col to int datatype

df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)

df.info()

# (Successfully converted)
```

```python
# Since we have Derived feature from Arrival time col so we dont need it anymore. So dropping it

df.drop('Arrival_Time', axis = 1, inplace = True)

df.head()

# Successfully removed
```

#### 3rd - On Departime time feature

```python
# Doing the same process which we did in arrival time feature
```

```python
# First splitting the hour and time into different columns

df['Departure_Hour'] = df['Dep_Time'].apply(lambda x:x.split(':')[0]) # Splitting using lambda
df['Departure_Minute'] = df['Dep_Time'].str.split(':').str[1] #done splitting normally

df.head()

# (Successfully added)
```

```python
# Now converting object dtype of Departure Hour and mintute into int datatype

df['Departure_Hour'] = df['Departure_Hour'].astype(int)
df['Departure_Minute'] = df['Departure_Minute'].astype(int)

df.info()

# (Successfully converted)
```

```python
# Since we have Derived feature from Dep_Time col so we dont need it anymore. So dropping it

df.drop('Dep_Time', axis = 1, inplace = True)

df.head()

# Successfully removed
```

#### 4th - On Total_Stops feature

```python
df['Total_Stops'].unique()

# Now since here is 'stop' word mentioned so we need to drop it, would be better for our model
```

```python
# Its also showing 'nan' as unique value so lets see which records have nan value for Total_Stops feature

df[df['Total_Stops'].isnull()]

# (As here we can see theres only 1 record so we can either drop it OR we can also guess like there is certainly not a direct
#  flight from Delhi to Cochin so it must have 1 stop.)

# So replacing 'nan' with 1 and dropping all 'stop/stops' word from this feature
```

```python
# Replacing

df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})

df['Total_Stops'].head()

#(Successfully replaced)
```

```python
# Dropping Route feature (as it's not really required for our model means it won't be thaaat impactful for our model) 
# Because we already have Source, Destination and also Total stops feature 

df.drop('Route', axis = 1, inplace = True)
```

```python
df.head(3)

# (Successfully removed Route)
```

#### 5th - On Duration column

```python
# Like Arrival time and Dep Time feature we dont have ':' in Duration. So splitting it based on 'h' and then 'm'

# df['Duration'].str.split(' ').str[0].str.split('h').str[0]
# Storing above loc in duation_hour

df['Duration_Hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df.info()

# Now need to convert Duration Hour dtype into int
```

```python
# Converting 
# df['Duration_Hour'] = df['Duration_Hour'].astype(int)

# Above loc(line of code) wont run cause there something '5m' in our column. So we need to check it
```

```python
df[df['Duration_Hour']=='5m']

# Now from Mumbai to Hyd their are 2 stops so its not possible to have the duration of 5m for those flights (5 minutes)
# Its better to drop these records
```

```python
# Dropping these records row wise as we already have the specified rows mentioned

df.drop(6474, axis = 0, inplace = True)
df.drop(2660, axis = 0, inplace = True)
```

```python
df[df['Duration_Hour']=='5m']

# (As is can be seen now there are no records for '5m' in duration hour)
# (Successfully removed)
```

```python
# Again Converting 

df['Duration_Hour'] = df['Duration_Hour'].astype(int)

# (Successfully converted)
```

```python
# Now dealing with minutes thing in duration feature
# Need to add col same as like we did for Duration_hour named Duration_minute

#df['Duration'].apply(lambda x:x.split('m')[0])
#df['Duration'].str.split(' ').str[1].str.split('m').str[0]

df['Duration_Minute'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0] 
df.head()
```

```python
df.info()

# (Now will convert duration minute to int )
```

```python
# Converting obj to int

# df['Duration_Minute'] = df['Duration_Minute'].astype(int)

# Won't be able convert to int cz their are nan values in duration minute
```

```python
# df[df['Duration_Minute'] == 'NaN']
# df[df['Duration_Minute']=='5m']
# df[df['Duration_Minute'].isnull()]
# df[df['Duration_Minute'] == 'NaN']
#df['Duration_Minute'] = df['Duration_Minute'].map({'NaN':0})
#df['Duration_Minute'].replace(np.nan,'0',regex = True)

df['Duration_Minute'] = df['Duration_Minute'].replace(np.nan,'0',regex = True)
df.head()
```

```python
# Now will convert duration minute to int 

df['Duration_Minute'] = df['Duration_Minute'].astype(int)

df.info()

# (Successfully converted)
```

```python
# Dropping Duration column because we already have derived duration minute and duration hour from it

df.drop('Duration', axis = 1, inplace = True)
```

```python
df.head(2)

# (Successfully removed)
```

```python
df.info()

# (Sucessfullly converted all the datatypes to either int or float.)
```
